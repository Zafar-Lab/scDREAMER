Building model

```python
data_path = "/home/ajita/Documents/data_integration/Lung/Lung_atlas_public.h5ad"
#data_path = "/home/ajita/Documents/data_integration/Lung/Lung_NA_0.5.h5ad"
```

```python
name = "Lung"

"""
give cell type NA for semi-supervised...
"""

batch_key_dic = {'Immune_Human' : 'batch',
                 'Lung' : 'batch',
                 'Pancreas' : 'tech',
                 'Human_Mouse' : 'batch',
                 'Human_Retina': "Batch"
                 
                }
# Below dictionary is an input to model for available cell types.
cell_type_key_dict = {'Immune_Human' : 'final_annotation',
                 'Lung' : 'cell_type',
                 'Pancreas' : 'celltype',
                 'Human_Mouse' : "celltype",
                 "Human_Retina":"Subcluster"
                    } 
    
plot_cell_type_dict = {
                    "Healthy_Heart" : "celltype",
                    'Immune_Human' : 'final_annotation', #'final_annotation',
                 'Lung' : 'cell_type',
                 'Pancreas' : 'celltype',
                 'Human_Mouse' : "celltype",
                 "Human_Retina":"Subcluster"
                    }  
```

```python

import src_shuffle.model_scdreamer_semisupervised as model

run_config = tf.ConfigProto()

run_config.gpu_options.per_process_gpu_memory_fraction = 0.333
run_config.gpu_options.allow_growth = True

with tf.Session(config = run_config) as sess:

    dreamer = model.scDREAMER(
        sess,
        epoch = 300,
        dataset_name = data_path,
        batch = batch_key_dic[name],
        cell_type = cell_type_key_dict[name],
        plot_cell_type = plot_cell_type_dict[name],
        name = name
        )

    dreamer.train_cluster()
    
```
